```python
def pattern(img_1,x,y):
    img_2=img_1.copy()
    for m in range(x-25,x+26):
        for n in range(y-25,y+26):
            """
            for k in range(3):
                if ((m-x)**2)+((n-y)**2)<=625:
                    img_2[n][m][k]=255
            """
            if ((m-x)**2)+((n-y)**2)<=625:
                    img_2[m][n]=255
    
    return img_2
```
